```python
from konlpy.tag import Mecab
mecab = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")
#final = mecab.nouns(case)
#final = [i for i in final if len(i) != 1]
#print(final)

relation = [[] for i in range(15)]

for i in xlist:
    tmp_mecab = []
    for j in case:
        if i in j:
            tmp_mecab = mecab.nouns(j)
            for k in tmp_mecab:
                if k != i and len(k) != 1:
                    relation[xlist.index(i)].append(k)
                
for i in relation:
    print(i)
```

```python
start = 0
slist3 = []
nlist3 = []
xlist3 = []
ylist3 = []

while start != 15:
    relation_c = Counter(relation[start]).most_common(5)
    for i in range(5):
        slist3.append(xlist[start])
        nlist3.append(ylist[start])
        xlist3.append(relation_c[i][0])
        ylist3.append(relation_c[i][1])
    start += 1
```

```python
import pandas as pd

numbers = []
for i in range(len(slist3)):
    numbers.append(i+1)

df2 = pd.DataFrame(numbers, columns = ['No'])
df2['Top_Word'] = slist3
df2['Number'] = nlist3
df2['Relation_Words'] = xlist3
df2['Numbers'] = ylist3
```
