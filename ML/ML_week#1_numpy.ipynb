# **기본 연산**

```python
# 더하기
data = np.array([[1, 2], [3, 4], [5, 6]])
ones_row = np.array([[1, 1]])
data + ones_row
```

```python
# 전체 array에서 최대값
data.max()
```

```python
# 전체 array에서 최소값
data.min()
```
