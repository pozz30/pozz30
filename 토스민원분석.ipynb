```python
from bs4 import BeautifulSoup
from selenium import webdriver
url = 'https://toss.im/every-moment'
# 이게 pc에 설치되어있는 chrome을 제어하는거라서 그냥 실행시키면 창이 나오고 
# 입력되고 하는게 다 보인다.
# headless option을 주어서 background에서 실행이 되도록 하자
driver_options = webdriver.ChromeOptions()
driver_options.add_argument("headless")
driver = webdriver.Chrome()
driver.get(url)
soup = BeautifulSoup(driver.page_source,'html.parser')
soup4 = soup.select_one('div.css-1cn7pka.e102nrhm0')
# 잊지말고 동작이 끝나면 driver을 종료할것!
driver.quit()
z = soup4.text
```

```python
# 동적 홈페이지 관련 크롤링 class 찾기가 애매해서 모든 class 크롤링 후 합하기
tosstext = a + b + c + z
tosstext = tosstext.split('번째 불편함')

tmp1 = []
tmp2 = []
# Number 및 TEXT 분류 후 리스트화
for idx,x in enumerate(tosstext[1:]):
    if x not in tmp2:
        tmp1.append(idx+1)
        tmp2.append(x[:-5])
print(tmp1, tmp2)
```

```python
# MYSQL Data 불러오기 (TOSS_VOC Table)
import pymysql
tdb = pymysql.connect(host='localhost',
user = 'root', password='0220vnfma!!', db = 'TOSS_TP',charset = 'utf8mb4')

cursor = tdb.cursor()

sql = "SELECT * FROM TOSS_VOC"
cursor.execute(sql)

case = []
result = cursor.fetchall()
for record in result:
    case.append(record[1])
tdb.close()
```

```python
case = '. '.join(case)
print(case)
```

```python
from konlpy.tag import Mecab
mecab = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")
final = mecab.nouns(case)
final = [i for i in final if len(i) != 1]
print(final)
```

```python
import numpy as np
from PIL import Image
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

counts = Counter(final)

#wc = WordCloud(font_path='malgun', width=800, height=400, scale=2.0, max_font_size=250)
#gen = wc.generate_from_frequencies(counts)
img = Image.open('책_검은배경.jpg')
img_array = np.array(img)

wc = WordCloud(font_path='malgun', width=600, height=300, scale=2.0, max_font_size=300, mask=img_array)
gen = wc.generate_from_frequencies(counts)

plt.figure(figsize=(15, 10))
plt.imshow(gen)
plt.axis('off')
plt.savefig('토스민원분석.jpg', bbox_inches='tight')
plt.show()
```
